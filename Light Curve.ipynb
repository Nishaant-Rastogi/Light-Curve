```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel("lco-2002_kl6.xlsx")
```

```python
print(dataset)
```

```python
dataset["Period"] = [4.61]*1007
dataset["Rotations"] = [4.61/24]*1007
dataset["Phase"] = dataset["Date"]/dataset["Rotations"] - (dataset["Date"]/dataset["Rotations"]).astype(int)
print(dataset)
```

```python
mag = dataset.iloc[:,6].values
date = dataset.iloc[:,3].values
plt.scatter(date,mag)

plt.show
```
